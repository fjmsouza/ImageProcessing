ABRIR, LER E MOSTRAR IMAGEM:

```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import cv2

def bgr2gray(bgr):
    return np.dot(bgr[...,:3], [ 0.1140, 0.5870, 0.2989 ])


# Read BGR image
input_image = cv2.imread('pout.bmp')
dimension = (input_image.shape[0] * input_image.shape[1])

print("Original image shape:",input_image.shape)
print("The image consists of %i pixels" % dimension)

img = input_image.copy()

#converting the BGR image to grayscale
gray = bgr2gray(img)

#rounding the elements
gray = np.round_(gray)
print("Gray image shape", gray.shape)

plt.subplot(1,3,1)
plt.imshow(input_image)
plt.title('Original (via 3 channels: R, G and B)')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gray)
plt.title('"True Gray" (1 channel)')
plt.axis('off')

```

CONTABILIZANDO PIXELS DA IMAGEM "TRUE GRAY" PARA CADA INTENSIDADE PRESENTE NA IMAGEM:

```python
#parcial histogram(counting pixels for each image's intensity
#through a dictionary where each key/value corresponds to intensity/quantity of pixels with this intensity
unique, counts = np.unique(gray, return_counts=True)
parcial_histogram = dict(zip(unique, counts))
print(parcial_histogram)
#separating the keys and values in lists
names = list(parcial_histogram.keys())
values = list(parcial_histogram.values())

plt.bar(range(len(parcial_histogram)), values, tick_label=names)
plt.show()
```

**CALCULANDO, DE FATO, O HISTOGRAMA DA IMAGEM "TRUE GRAY"**

CRIANDO VETOR HISTOGRAMA_ORIGINAL VAZIO:

```python
intensity_levels = 256
histograma_original = [0]*intensity_levels
print(histograma_original)
```

CALCULANDO(POPULANDO) VETOR HISTOGRAMA_ORIGINAL:

```python
for i in range(len(histograma_original)):
  if i in parcial_histogram:
    histograma_original[i] = parcial_histogram.get(i)
  else:
    histograma_original[i] = 0

print(histograma_original)

# Horizontal Bar Plot
plt.bar(range(256), histograma_original)
```

CRIANDO E CALCULANDO A FUNÇÃO DE DISTRIBUIÇÃO ACUMULADA (*fda*):

```python
fda = [0]*256
print(fda)

# Calculating the general coeficient for PDF
coeficient= (intensity_levels-1)/dimension

# Performing elementwise operation on all elements of the array to obtain the
# histogram equalization equation
fda[0] = histograma_original[0]

for i in range(1,len(fda)):
  if (i >= 1):
    fda[i] = fda[i-1] + histograma_original[i]

fda = [i * coeficient for i in fda]

#rounding element to integer
fda = np.round_(fda)


plt.plot(fda)

```

EQUALIZANDO IMAGEM "TRUE GRAY"

```python
# mapping each pixel intensity in the x axis in fda to the correspondent intensity in y axis
gray_eq= np.copy(gray)
for x in range(gray.shape[0]):
  for y in range(gray.shape[1]):
    gray_eq[x,y] = fda[int(gray[x,y])]

plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.title('"True Gray"')
plt.imshow(gray)
plt.subplot(1, 2, 2) # row 1, col 2 index 2
plt.title('"True Gray" Equalizada')
plt.imshow(gray_eq)
```

# COMPARAÇÃO ENTRE HISTOGRAMAS

OS ARTEFATOS A SEGUIR SÃO ORIUNDOS DOS PASSOS ANTERIORES:

```python
#void vector
histograma_eq = [0]*256

#parcial histogram(counting pixels for each image's intensity
unique, counts = np.unique(gray_eq, return_counts=True)
parcial_histograma_eq = dict(zip(unique, counts))
names = list(parcial_histograma_eq.keys())
values = list(parcial_histograma_eq.values())

for i in range(len(histograma_eq)):
  if i in parcial_histograma_eq:
    histograma_eq[i] = parcial_histograma_eq.get(i)
  else:
    histograma_eq[i] = 0

#showing both histograms
plt.bar(range(256), histograma_original)
plt.show()
plt.bar(range(256), histograma_eq)
plt.show()
```

ALGORITMO 1:

*MEAN GRAY LEVEL*

```python
def meanGrayLevel(histogram):

  #obtaining the least intensity
  i=0;
  while (histogram[i]==0):
    i=i+1

  gray_minimum = i

  #obtaining the greater intensity, backwards
  i=255
  while (histogram[i]==0):
    i=i-1

  gray_maximum = i

  #calculating the mean value between the extremes
  mean_gray = round((gray_minimum + gray_maximum)/2)

  print("O nível de cinza médio é",mean_gray )

  return mean_gray
```

*BINARIZAÇÃO*

```python
# in this function is passed 3 parameters: the image, its histogram and cut point (optional)
def binarization(image, histograma, cut = None):

  image_bin= np.copy(image)
  if cut == None:
    cut = meanGrayLevel(histograma)
  for x in range(image_bin.shape[0]):
    for y in range(image_bin.shape[1]):
      # in fact, binarization process through the threshold ...
      if (image[x,y]>= cut):
        image_bin[x,y] = 255
      else:
        image_bin[x,y] = 0

  return image_bin
```

_PLOTANDO BINARIZAÇÃO DAS IMAGENS SEM E COM EQUALIZAÇÃO, RESPECTIVAMENTE_

```python

plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.title('"True Gray" Binarizada')
plt.imshow(binarization(gray, histograma_original))
plt.subplot(1, 2, 2) # row 1, col 2 index 2
plt.title('"True Gray" Equalizada Binarizada ')
plt.imshow(binarization(gray_eq, histograma_eq))
```

ALGORITMO 2:

*TWO PEAKS*

```python
def minimumBetween2Peaks(histogram):

  #initializing the variables
  i=0
  first_peak = 0
  first_peak_index = 0
  #obtaining the first peak after sweep the entire vector
  for i in range(len(histogram)):
    if histogram[i] >= first_peak:
      first_peak = histogram[i]
      first_peak_index = i

  #reinitializing the variables
  i = 255
  second_peak = 0
  second_peak_index = 0
  flag_stop = False
  #obtaining the second peak, backwards until find the decay
  while not flag_stop:
    if (histogram[i] >= second_peak):
      second_peak = histogram[i]
      second_peak_index = i
      #decrement the index
      i = i - 1
    else:
      if (histogram[i] != 0):
        # stop the seaching if element decay and different of 0
        flag_stop = True

  #searching for the minimum between the peaks
  #to facilitate this searching, slicing the histogram
  histogram_sliced = histogram[first_peak_index:second_peak_index]
  #initializing the minimum gray
  minimum = max(histogram_sliced);
  #sweep entire slice
  for i in range(len(histogram_sliced)):
    #seek index of the minimum value not null
    if (histogram_sliced[i] != 0) and (histogram_sliced[i]<minimum) :
       minimum_index = i
  #restore the index of minimum value in histogram
  minimum_gray_level = minimum_index + first_peak_index
  print("O nível de cinza mínimo entre dois picos é",minimum_gray_level )
  return minimum_gray_level
```
